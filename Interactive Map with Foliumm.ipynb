```python
import folium

# Create a map centered around a specific location
m = folium.Map(location=[37.7749, -122.4194], zoom_start=12)  # San Francisco

# Add a marker
folium.Marker(
    location=[37.7749, -122.4194],
    popup='San Francisco',
    icon=folium.Icon(color='blue')
).add_to(m)

# Add a circle marker
folium.CircleMarker(
    location=[37.7749, -122.4194],
    radius=10,
    color='red',
    fill=True,
    fill_color='red'
).add_to(m)

# Save the map
m.save('map.html')
```

```python
import folium

# Example data: latitude, longitude, and outcome
launches = [
    {'location': [34.6367, -120.4328], 'outcome': 'Success'},
    {'location': [28.5623, -80.5774], 'outcome': 'Failure'},
    {'location': [39.0474, -76.6715], 'outcome': 'Success'},
    {'location': [37.7749, -122.4194], 'outcome': 'Success'},
    {'location': [43.6426, -79.3871], 'outcome': 'Failure'}
]

# Initialize map
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Add markers with color-coding based on outcome
colors = {'Success': 'green', 'Failure': 'red'}

for launch in launches:
    folium.Marker(
        location=launch['location'],
        popup=f"Outcome: {launch['outcome']}",
        icon=folium.Icon(color=colors[launch['outcome']])
    ).add_to(m)

# Save the map as an HTML file
m.save('launch_outcomes_map.html')
```

```python
import folium
from geopy.distance import geodesic

# Coordinates of the selected launch site
launch_site = [34.6367, -120.4328]  # Example: Vandenberg Space Force Base

# Proximities (example coordinates)
railway = [34.6390, -120.4250]
highway = [34.6310, -120.4200]
coastline = [34.6270, -120.4350]

# Create map centered around the launch site
m = folium.Map(location=launch_site, zoom_start=14)

# Add launch site marker
folium.Marker(
    location=launch_site,
    popup='Launch Site',
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(m)

# Add railway marker
folium.Marker(
    location=railway,
    popup='Railway',
    icon=folium.Icon(color='gray', icon='train')
).add_to(m)

# Add highway marker
folium.Marker(
    location=highway,
    popup='Highway',
    icon=folium.Icon(color='orange', icon='road')
).add_to(m)

# Add coastline marker
folium.Marker(
    location=coastline,
    popup='Coastline',
    icon=folium.Icon(color='lightblue', icon='info-sign')  # Using 'lightblue' instead of 'cyan'
).add_to(m)

# Calculate and display distances
def add_distance_marker(start, end, label):
    distance = geodesic(start, end).kilometers
    folium.Marker(
        location=[(start[0] + end[0]) / 2, (start[1] + end[1]) / 2],
        popup=f'{label} Distance: {distance:.2f} km',
        icon=folium.Icon(color='purple')
    ).add_to(m)

add_distance_marker(launch_site, railway, 'Launch Site to Railway')
add_distance_marker(launch_site, highway, 'Launch Site to Highway')
add_distance_marker(launch_site, coastline, 'Launch Site to Coastline')

# Save the map as an HTML file
m.save('launch_site_proximities_map.html')
```
